```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data_path = 'Final_project/depression_data.csv'
df = pd.read_csv(data_path)
```

```python
print("Dataset Overview:")
print(df.info())
print("\nFirst 5 Rows of the Dataset:")
print(df.head())
```

```python
missing_data = df.isnull().mean() * 100
print("\nMissing Values (%):")
print(missing_data)
```

```python
if "Age" in df.columns:
    bins = [0, 18, 25, 35, 45, 60, 100]
    labels = ['Under 18', '18-25', '26-35', '36-45', '46-60', '60+']
    df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
columns_to_remove = ['Name']
df.drop(columns=columns_to_remove, inplace=True, errors='ignore')
```

```python
print("\nCleaned Dataset Overview:")
print(df.info())
print("\nFirst 5 Rows of the Cleaned Dataset:")
print(df.head())
```

```python
output_path = 'Final_project/cleaned_depression_data.csv'
df.to_csv(output_path, index=False)
print(f"\nCleaned dataset saved to {output_path}.")
```

```python
family_history_count = 111253
total_count = 413768

no_family_history_count = total_count - family_history_count

labels = ['Family History of Depression', 'No Family History']
sizes = [family_history_count, no_family_history_count]
colors = ['#ff9999', '#66b3ff']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Number of Individuals with Family History of Depression')
plt.axis('equal')
plt.show()
```

```python
percentage_mental_illness = 30.41

labels = ['Mental Illness', 'No Mental Illness']
sizes = [percentage_mental_illness, 100 - percentage_mental_illness]
colors = ['#ff9999','#66b3ff']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Individuals with Mental Illness')
plt.axis('equal')
plt.show()
```

```python
both_conditions_count = 39347
mental_illness_count = 125825
substance_abuse_count = 128888

categories = ['Mental Illness Only', 'Substance Abuse Only', 'Both Conditions']
values = [mental_illness_count, substance_abuse_count, both_conditions_count]

plt.bar(categories, values, color=['#ff9999', '#66b3ff', '#99ff99'])

plt.title('Proportion of Individuals with Mental Illness and Substance Abuse')
plt.ylabel('Count')

plt.show()
```
